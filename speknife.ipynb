There are two input block options for using Speknife:

1.   Simple Correction: This option allows you to perform a basic correction where you can select the plot options displayed below.

2.   Complete Correction: This option includes uncertainty calculations and provides the option to estimate the tube kV value, which is displayed at the end of the notebook for users who want to explore this feature.

To run the correction, you need to:

*   Click the Run button for the Drive mount,
*   Run the input.txt configuration
*   Finally, click the Speknife button.


For full access to the Python code, it is recommended to open the developer version. You can contribute by making your own improvement suggestions on GitHub (https://github.com/matheusrebe/Speknife).
